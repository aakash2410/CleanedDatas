```python
df = pd.read_csv('employee.csv')
```

```python
df['BusinessTravel'].unique()
```

```python
for i in range(df["BusinessTravel"].count()):
    if df["BusinessTravel"][i] == 'Travel_Rarely':
        df["BusinessTravel"][i] = "1"
    elif df["BusinessTravel"][i] == 'Travel_Frequently':
        df["BusinessTravel"][i] = '2'
    elif df["BusinessTravel"][i] == 'Non-Travel':
        df["BusinessTravel"][i] = '0'
```

```python
df['BusinessTravel'].unique()
```

```python
df['BusinessTravel']=df['BusinessTravel'].astype(dtype=int)
```

```python
df.info()
```

```python
df['OverTime'].unique()
```

```python
for i in range(df['OverTime'].count()):
    if df['OverTime'][i] == 'Yes':
        df['OverTime'][i] = '1'
    elif df['OverTime'][i] == 'No':
        df['OverTime'][i] = '0'
        
df['OverTime'] = df['OverTime'].astype(dtype = int)
```

```python
df["OverTime"].unique()
```

```python
df.info()
```

```python
for i in range(df['Attrition'].count()):
    if df['Attrition'][i] == 'Yes':
        df['Attrition'][i] = '1'
    elif df['Attrition'][i] == 'No':
        df['Attrition'][i] = '0'
        
df['Attrition'] = df['Attrition'].astype(dtype = int)
df['Attrition'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
OHE = OneHotEncoder()
```

```python
temp = OHE.fit_transform(df[['EducationField']]).toarray()
```

```python
temp1 = OHE.fit_transform(df[['Department']]).toarray()
```

```python
temp2 = OHE.fit_transform(df[['MaritalStatus']]).toarray()
temp3 = OHE.fit_transform(df[['Gender']]).toarray()
temp4 = OHE.fit_transform(df[['JobRole']]).toarray()
```

```python
temp, temp1, temp2, temp3, temp4
```

```python

temp = pd.DataFrame(temp, columns = df['EducationField'].unique())
```

```python
temp1 = pd.DataFrame(temp1, columns = df['Department'].unique())
temp1
```

```python
temp2 = pd.DataFrame(temp2, columns = df['MaritalStatus'].unique())
temp2
```

```python
temp3 = pd.DataFrame(temp3, columns = df['Gender'].unique())
temp3
```

```python
temp4 = pd.DataFrame(temp4, columns = df['JobRole'].unique())
temp4
```

```python
temp_new = pd.concat([temp,temp1,temp2,temp3,temp4], axis = 1)
```
